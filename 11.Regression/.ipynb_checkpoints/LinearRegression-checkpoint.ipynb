### 회귀분석
* 두 변수 사이의 연관성을 알아내는 통계 방법을 회귀분석이라 한다

ex) 매출액에 영향을 미치는 변수는 무엇인가?

```python
import matplotlib as plt
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# 공부시간 : 독립변수
# 시험점수 : 종속변수
x = [[2], [4], [6], [8], [10]]
y = [[81], [93], [90], [97], [100]]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
result = model.predict([[7]]) # 7: 학습시간
print(f'예상점수: {result}')
```

평균값 하고 중앙값을 보면 2정도 차이 난다
분산값(std) 지점 값을 보면 0 이다. --> 분석 가치가 없다는걸 알 수 있다.

```python
# 지점 컬럼 삭제 
# seoul.drop('지점', axis=1, inplace=True) #inplace=True 사용할 때는 신중하게 사용한다. , axis=1  세로로 지운다
seoul.head()
```

```python
# 컬럼이름 변경
seoul.columns=['날짜', '평균기온', '최저기온', '최고기온']
seoul.columns
```

```python
#결측치 확인
seoul.isnull().sum()
```

```python
seoul.dropna(subset=['최고기온'], axis=0, inplace=True)
seoul.info()
```

```python
#파생변수
#컬럼 추가 (년, 월, 일)
seoul['년도'] = seoul['날짜'].dt.year
seoul['월'] = seoul['날짜'].dt.month
seoul.head()
```

```python
conditions = (seoul['날짜'].dt.month==8) & (seoul['날짜'].dt.day ==15)
seoul0815 = seoul[conditions]
seoul0815.head(10)
```

```python
seoul0815.info()
```

```python
seoul0815.isnull().sum()
```

```python
# plt.re('font', famaily='NanumGothic') #for window
plt.rcParams['font.family'] = 'NanumGothic'

print(plt.rcParams['font.family'])
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15, 7))
x = seoul0815[['년도']]
y = seoul0815['평균기온']
plt.xlabel('년도')
plt.ylabel('평균기온')
plt.scatter(x, y)
plt.show()
```

```python
import seaborn as sns
sns.jointplot(x='년도', y='평균기온', kind='reg', data=seoul)
```

```python
# 단순 선형 회귀
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = seoul0815[['년도']]   #data 차원 맞추기
print(X.shape, seoul0815['년도'].shape)
y = seoul0815['평균기온']
model.fit(X,y)
```

```python
result = model.predict([[2024]])
print('2022년도 평균기온 예측 값:', result)
```

```python
# coed_ : 기울기(a,w)
#intercept_ : 절편, 바이어스(b)
print('기울기:', model.coef_)
print('절편:', model.intercept_)
```

```python
seoul.info()
```

```python
# 다중 선형 회귀
model = LinearRegression()
X = seoul[['년도', '최저기온', '최고기온']]
y = seoul['평균기온']

model.fit(X,y)
```

```python
result = model.predict([[2022, 24, 33]])
print(result)
```

평균온도는 28도로 예측했다.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape, y_train.shape, X_test, y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(y_test[:1].values.shape)
print(y_test[:1].values.reshape(-1,1).shape)
```

```python
plt.plot(model.predict(X_test[:50]), label='predic')
plt.plot(y_test[:50].values.reshape(-1,1), label='real temp')
plt.legend()
```

```python
plt.figure(figsize=(15, 7))
y_pred = model.predict(X_test)
plt.scatter(y_test, y_pred, alpha=0.4)
plt.show()
```

모델 성능 평가
* MAE :  평균 절대 오차 <= 적은게 좋음  
* MSE : 평균 제곱 오차 <= 적은게 좋음  
* RMSE : 제곱 단위를 원래 단위로 본귀  
* R2 : 1에 가까울 수록 좋음. X  변수들이 Y 변수에 미치는 영향   
       X 변수들이 평균 기온값 변동의 99%를 설명 할 수 있다는 의미
    

```python
# R2
print(model.score(X_test, y_test))
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print('mae: ', mae)
print('mse:', mse)
print('rmse:', rmse)
```
